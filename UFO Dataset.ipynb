```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.info()
```

```python
ufo.State.value_counts().plot()
```

```python
ufo.timestamp.dt.hour.value_counts().sort_index().plot()
```
